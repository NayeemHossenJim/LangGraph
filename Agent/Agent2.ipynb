```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    age: int
    details: str
```

```python
def multiple_inputs(state: AgentState) -> AgentState:
    """A node that takes multiple inputs and produces a new output."""
    state["details"] = f"{state['name']} is {state['age']} years old."
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("Multiple Inputs", multiple_inputs)
graph.set_entry_point("Multiple Inputs")
graph.set_finish_point("Multiple Inputs")
app = graph.compile()
app
```

```python
result = app.invoke({"name": "Alice", "age": 30})
result["details"]
```
