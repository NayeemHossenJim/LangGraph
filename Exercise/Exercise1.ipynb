```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name : str
```

```python
def complement_agent(state: AgentState)-> AgentState:
    """A simple agent that takes a name and responds with a complement."""
    state['name'] = "Hey, "  + state['name'] + " You are doing great!"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("complement", complement_agent)
graph.set_entry_point("complement")
graph.set_finish_point("complement")
app = graph.compile()
app
```

```python
result = app.invoke({"name": "Alice"})
result["name"]
```
